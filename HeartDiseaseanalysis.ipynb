```python
##Check chest pain type  cp denotes chest pain
##type of chest pain
## value 0: typical angina
## values 1 : atypical angina
## value 2:non-anginal pain
##value 3: asymptomatic pain
```

```python
df['cp'].value_counts()
```

```python
plt.figure(figsize=[8,4])
sns.countplot(x='cp',data=df)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic pain'])
plt.xlabel('Chest pain', fontsize=12)
plt.ylabel('No of cp',fontsize=12)
plt.title("Number and types of chest pain",fontsize=18)
plt.show()
```
